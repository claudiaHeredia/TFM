Como hemos podido observar "age" viene como [[60.0, 'year']] y "gender" como M/F.  

El siguiente paso ser√° normalizar "age" y "gender":

- age_col_norm (int) 
     
    "[[60.0, 'year']]"  -> 60
    "60" / "60.0"       -> 60
    Devuelve int o None si no encuentra un valor razonable (0<edad<130).

- sex_col_norm (male/female)

    Mapea variantes a {'male','female'}.
    Acepta: 'M','F','male','female','man','woman','hombre','mujer', etc.

para poder comparar con lo que se extraiga del texto.

```python
import re

def parse_age_col(val:str):
    if not val or not isinstance(val, str):
        return None
    m = re.search(r'([0-9]+(?:\.[0-9]+)?)',val)
    if not m:
        return None
    try:
        num = float(m.group(1))
        if 0 < num < 130:
            return int(round(num))
    except:
        return None
    return None

def normalize_gender_col(val:str):
    if not val:
        return None
    v = val.strip().lower()
    if v in {"m", "male", "man", "hombre"}:
        return "male"
    if v in {"f", "female", "woman", "mujer"}:
        return "female"
    return None

#Creamos las columnas normalizadas:

df["age_col_norm"] = df.get("age", "").apply(parse_age_col)
df["sex_col_norm"] = df.get("gender", "").apply(normalize_gender_col)

#Vista
try:
    display(df[["age", "age_col_norm", "gender", "sex_col_norm"]].head(5))
except NameError:
    print(df[["age","age_col_norm","gender","sex_col_norm"]].head(5))
```
